```python
def lietke_chuoinhiphan(n):
    danh_sach = []
    #bat_dau = 1
    for i in range(0, 2**n): 
      gia_tri = i
      chuoi_np = bin(gia_tri).replace('0b','')
      while (len(chuoi_np) < n):
        chuoi_np = '0' + chuoi_np
      danh_sach.append(chuoi_np)  #them chuoi vao kq
        #print (chuoi_np)
    return danh_sach
        
```
