```python
def lietke_chuoinhiphan(n):
    danh_sach = []
    #bat_dau = 1
    for i in range(0, 2**n): 
      gia_tri = i
      chuoi_np = bin(gia_tri).replace('0b','')
      while (len(chuoi_np) < n):
        chuoi_np = '0' + chuoi_np
      danh_sach.append(chuoi_np)  #them chuoi vao kq
        #print (chuoi_np)
    return danh_sach
        
```

```python
def kiemtra_chuoi(chuoi_dang_chon, danhsach, gioihan):
  tong_thoigian = 0
  for j in range(0, len(chuoi_dang_chon)):
    if chuoi_dang_chon[j] == '1':
      tong_thoigian = tong_thoigian +thoigian[j]
  ketqua = True  #nghia la ok
  if tong_thoigian > gioihan:
    ketqua = False # khong duoc
  return ketqua, tong_thoigian
```

```python
import matplotlib.pyplot as plt
import numpy.random as rnd
from matplotlib.patches import Ellipse
from scipy import rand 
NUM = 250
# xy: center of ellis, w: duong kinh ngang, h: chieu cao, angle: goc xoay
ells = [Ellipse (xy = rnd.rand(2)* 10, width= rnd.rand(),
                 height= rnd.rand(), angle= rnd.rand()* 360)
        for i in range (NUM)]

fig = plt.figure(0)
# tao ra 1 luoi gom co 1 dong, 1 cot, va vi tri hien tai la o 1
ax = fig.add_subplot(111, aspect = 'equal')
for e in ells:
  ax.add_artist(e)
  e.set_clip_box(ax.bbox)
  e.set_alpha(rnd.rand())
  e.set_facecolor(rnd.rand(3))

ax.set_xlim(0, 10)    
ax.set_ylim(0, 10)  
plt.show()
```
